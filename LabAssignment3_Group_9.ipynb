### Solving the primal problem

### Primal problem 
$$
\begin{aligned}
\min_{w \in \mathbb{R}^n,\; b \in \mathbb{R}} \quad
& \frac{1}{2}\lVert w \rVert_2^2 \\
\text{s.t.} \quad
& 1 - y^{(i)}\bigl(w^\top x^{(i)} + b\bigr) \le 0,
\quad \forall i \in \{1,2,\ldots,N\}.
\end{aligned}
$$

```python
def build_dataset(A, B):
    x = np.vstack((A, B))
    y = np.hstack((np.ones(len(A)), -np.ones(len(B))))
    return x, y

def solve_dual_svm(x, y):
    N = x.shape[0]
    lam = cp.Variable(N)

   
    Q = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            Q[i, j] = y[i] * y[j] * np.dot(x[i], x[j])

    objective = cp.Maximize(
        cp.sum(lam) - 0.5 * cp.quad_form(lam, cp.psd_wrap(Q))
    )

    constraints = [
        lam>=0,
        cp.sum(cp.multiply(lam, y)) == 0
    ]


    cp.Problem(objective, constraints).solve()
    return lam.value


def recover_primal(x, y, lam, tol=1e-6):
    w = np.sum((lam * y)[:, None] * x, axis=0)

    sv_idx = np.where(lam > tol)[0][0]
    b = (1 / y[sv_idx]) - np.dot(w, x[sv_idx])

    return w, b


def verify_kkt_conditions(x, y, lam, w, b, tol=1e-6):
    satisfied = True

    
    if not (np.all(lam >= -tol) and abs(np.sum(lam * y)) <= tol):
        print("Dual feasibility violated")
        satisfied = False

    
    margins = 1 - y * (x @ w + b)
    if not np.all(margins <= tol):
        print("Primal feasibility violated")
        satisfied = False

    
    if not np.all(np.abs(lam * margins) <= tol):
        print("Complementary slackness violated")
        satisfied = False

    
    w_check = np.sum((lam * y)[:, None] * x, axis=0)
    if not np.allclose(w, w_check, atol=tol):
        print("Stationarity violated")
        satisfied = False

    if satisfied:
        print("All KKT conditions are satisfied.")
    else:
        print("KKT conditions are NOT satisfied.")

    return satisfied



def run_svm_dual_with_kkt(A, B):
    x, y = build_dataset(A, B)
    lam = solve_dual_svm(x, y)
    w, b = recover_primal(x, y, lam)
    verify_kkt_conditions(x, y, lam, w, b)

    return lam, w, b

```

### The relaxed SVM problem

$$\min_{w \in \mathbb{R}^K, b \in \mathbb{R}, \xi \in \mathbb{R}^N} \quad \frac{1}{2}\|w\|_2^2 + C \sum_{i=1}^N \xi_i$$

$$\text{s.t.} \quad 1 - \hat{y}_i(w^\top \phi(\hat{x}_i) + b) \leq \xi_i, \quad \xi_i \geq 0, \quad \forall i \in \{1, 2, \dots, N\}$$

```python
def solve_soft_margin_svm_primal(A, B, C=1.0):
    x = np.vstack((A, B))
    y = np.hstack((np.ones(len(A)), -np.ones(len(B))))
    N, d = x.shape

    w = cp.Variable(d)
    b = cp.Variable()
    zetai = cp.Variable(N)

    objective = cp.Minimize(
        0.5 * cp.sum_squares(w) + C * cp.sum(zetai)
    )

    constraints = [
        1 - cp.multiply(y, x @ w + b) <= zetai,
        zetai >= 0
    ]

    problem = cp.Problem(objective, constraints)
    problem.solve()
    return w.value, b.value, zetai.value
```

```python
def plot_svm_hyperplanes(A, B, w, b, title="Relaxed SVM"):
    

    plt.figure(figsize=(8, 6))

    # Scatter data
    plt.scatter(A[:, 0], A[:, 1], c='r', marker='o', label='Class +1')
    plt.scatter(B[:, 0], B[:, 1], c='b', marker='s', label='Class -1')

    # Line range
    x_min = min(A[:, 0].min(), B[:, 0].min()) - 1
    x_max = max(A[:, 0].max(), B[:, 0].max()) + 1
    x_lin = np.linspace(x_min, x_max, 300)

    # Hyperplanes
    y_dec = -(w[0] * x_lin + b) / w[1]
    y_pos = -(w[0] * x_lin + b - 1) / w[1]
    y_neg = -(w[0] * x_lin + b + 1) / w[1]

    plt.plot(x_lin, y_dec, 'k-', linewidth=2, label='Decision Boundary')
    plt.plot(x_lin, y_pos, 'k--', alpha=0.7, label='Margin +1')
    plt.plot(x_lin, y_neg, 'k--', alpha=0.7, label='Margin -1')

    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title(title)
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

# Using the kernel method when the number of outliers is very high 

The given kernel
$$K(x, y) = e^{(-\gamma \|x - y\|_2^2)}$$

for $\gamma \in \{0.1, 1, 10\}$

```python
gamma_vals = [0.1,1,10]
```

```python
def gaussian_kernel(X_data,Y_data,C=1.0,gamma=0.1):
    alpha=cp.Variable(X_data.shape[0])
    K=np.zeros((X_data.shape[0],X_data.shape[0]))
    for i in range(X_data.shape[0]):
        for j in range(X_data.shape[0]):
            K[i,j]=np.exp(-gamma*np.sum((X_data[i]-X_data[j])**2))
    Y_values=Y_data.reshape(-1,1)
    Q=(Y_values@Y_values.T)*K
    objective=cp.Maximize(cp.sum(alpha)-0.5*cp.quad_form(alpha,cp.psd_wrap(Q)))
    constraints=[alpha>=0,
                 alpha<=C,
                 cp.sum(cp.multiply(alpha,Y_data))==0]
    prob=cp.Problem(objective,constraints)
    prob.solve()
    print(f"Problem Status: {prob.status}")
    if prob.status == 'optimal':
        print("Solved Using the Gaussian Kernel!")
        print(f"Alpha for gamma={gamma}:{alpha.value}")
    return alpha.value
```

## Plotting for the Dataset 3 

```python
def plot_grid(X_data,Y_data,alpha,gamma=0.1,C=1):
    epsilon = 1e-5
    indices_k = np.where((alpha > epsilon) & (alpha < C - epsilon))[0]
    
    k = indices_k[0]
    
    # calculating (w*)^T phi(x_k) using the kernel
    w_phi_xk = 0
    for i in range(len(alpha)):
        if alpha[i] > epsilon: # Only need support vectors
            dist_sq = np.sum((X_data[i] - X_data[k])**2)
            kernel_val = np.exp(-gamma * dist_sq)
            w_phi_xk += alpha[i] * Y_data[i] * kernel_val
    
    # b* = y_k - (w*)^T phi(x_k)
    bias = Y_data[k] - w_phi_xk


    res = 0.1
    x_range = np.arange(-2, 2 + res, res)
    y_range = np.arange(-2, 2 + res, res)
    
    #creating the grid
    xx, yy = np.meshgrid(x_range, y_range)
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    
    #predicting labels for all points in the grid
    Z = []
    for p in grid_points:
        sv_idx = np.where(alpha > 1e-5)[0]
        score = bias
        for idx in sv_idx:
            dist_sq = np.sum((X_data[idx] - p)**2)
            score += alpha[idx] * Y_data[idx] * np.exp(-gamma * dist_sq)
        
        Z.append(1 if score >= 0 else -1)
    
    Z = np.array(Z).reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    
    plt.pcolormesh(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.3, shading='auto')
    
    #plotting original datapoints
    plt.scatter(X_data[Y_data == 1, 0], X_data[Y_data == 1, 1], 
                c='blue', edgecolors='k', label='+1 labelled Datapoints')
    plt.scatter(X_data[Y_data == -1, 0], X_data[Y_data == -1, 1], 
                c='red', edgecolors='k', label='-1 labelled Datapoints')

    plt.title(f"Nonlinear SVM Classification (gamma={gamma})")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.legend()
    plt.show()
```
